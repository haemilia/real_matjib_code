```python
import pandas as pd
df = pd.read_csv("C:/medici_class/00.프로젝트/최종프로젝트 임시파일/gmaps_reviews_selenium_re.csv")
df
```

```python
df = df.drop_duplicates()
df
```

```python
# 1. 리뷰 내용이 공백 또는 NaN인 행들만 선택
empty_reviews = df[df['리뷰 내용'].isna() | (df['리뷰 내용'].str.strip() == '')]

# 동일한 작성자와 사업장명 기준으로 하나만 남기고 제거
empty_reviews_deduped = empty_reviews.drop_duplicates(subset=['작성자', '사업장명'])

# 공백 리뷰가 아닌 나머지 데이터
non_empty_reviews = df[~df.index.isin(empty_reviews.index)]

# 2. 리뷰 내용이 중복인 경우 하나만 남김
non_empty_reviews_deduped = non_empty_reviews.drop_duplicates(subset=['리뷰 내용'])

# 3. '고객님'이라는 단어가 포함된 리뷰 제거
filtered_reviews = non_empty_reviews_deduped[~non_empty_reviews_deduped['리뷰 내용'].str.contains('고객님', na=False)]

# 최종 결과: 공백 리뷰 중 하나만 남긴 것 + 중복 제거 및 '고객님' 필터링된 리뷰
final_df = pd.concat([empty_reviews_deduped, filtered_reviews], ignore_index=True)

# 결과 확인
final_df
```

```python
final_df.to_csv("cleaned_reviews.csv", index=False)
```

### 구글맵스 리뷰 라벨링 기준
1. 사업장 100개 랜덤 추출하여 라벨링 진행 
2. 리뷰내용 공백인 경우는 비광고성으로 분류
3. 비광고성 리뷰 : 1 / 광고성 리뷰 : 0 으로 라벨링  
4. 기준1과 기준2의 경우 광고성 리뷰로 분류, 주관적 판단인 기준3의 경우 2로 라벨링 후 팀원 크로스체크 이후 0으로 분류 예정
5. 플랫폼 특성상 광고성 리뷰의 비중이 높지 않음
6. 이미지로 홍보성 여부 판단하기 어려움으로 텍스트 위주로 판단 진행

**[광고성 리뷰 기준1]**  
* 리뷰에 광고성 단어 포함된 경우(협찬, 제공받아, 할인받아, 작성한, 체험단, 슈퍼멤버스, 리뷰이벤트 등)  
** 광고성 리뷰임이 확실함

**[광고성 리뷰 기준2]**  
* 리뷰 내용에 해시태그 포함된 경우  
** 해시태그로 검색하는 플랫폼이 아닌 만큼 해시태그 적용 시 광고일 가능성 높다고 판단  

**[광고성 리뷰 기준3]**  
* 리뷰 평점이 가게 전체 별점 대비 높은 경우 
* 매장의 풀네임이 리뷰내용에 추가된 경우
* 연남동+사업장명 방문했어요~, 연남동/홍대 맛집입니다~ 와 같은 뉘앙스의 내용 포함된 경우  
* 데이트 장소로 추천하는 경우, 매장이 깔끔하고 직원이 친절하다는 내용 포함된 경우  
* 구체적인 가격 명시된 경우  
** 주관적인 판단으로 광고라고 의심이 되는 경우, 위 기준에 적합할 시 광고성 리뷰로 분류  
** 번역체의 리뷰는 비광고성으로 판단

```python
df_s = pd.read_csv("C:/medici_class/00.프로젝트/최종프로젝트 임시파일/sampled_reviews.csv")
df_s
```

```python
keywords = ['협찬', '제공받아', '할인받아', '작성한', '체험단', '슈퍼멤버스', '리뷰이벤트', '할인']

# 리뷰 내용에 키워드 중 하나라도 포함되거나 '#' 기호가 포함된 경우 라벨을 0, 아니면 1
def label_review(text):
    if pd.isna(text):
        return 1
    if any(keyword in text for keyword in keywords) or '#' in text:
        return 0
    return 1

# 새 열 '라벨' 추가
df_s['라벨링'] = df_s['리뷰 내용'].apply(label_review)
df_s
```

```python
df_s.to_csv("labeled_reviews.csv", index=False)
```

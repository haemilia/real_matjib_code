## 카카오맵 일반음식점 상세보기 주소 크롤링
* 마포구 일반음식점 중 연남동 소재 일반음식점들만 추려냄
* 연남동 음식점들의 카카오맵 '상세보기' 주소 가져와 정리

```python
import pandas as pd
import numpy as np

# CSV 파일 불러오기
df = pd.read_csv('seoul_mapogu_general_restaurants.csv', encoding='cp949')
df
```

```python
## 전체 컬럼명 확인
df.columns
```

```python
## 필요한 컬럼만 선택해 작업용 DF 만들기
filtered_df = df[['사업장명', '소재지전체주소', '도로명전체주소', '좌표정보X(EPSG5174)', '좌표정보Y(EPSG5174)']]
filtered_df
```

```python
# '소재지전체주소' 기준으로 '연남동'이 포함된 행만 추출
filtered_df = filtered_df[filtered_df['소재지전체주소'].str.contains('연남동', na=False)]
filtered_df
```

```python
## 세 개의 컬럼 데이터(사업장명, x좌표, y좌표)가 동일한 경우 첫 번째 값만 유지하고 나머지는 중복으로 제거 
filtered_df = filtered_df.drop_duplicates(subset=['사업장명', '좌표정보X(EPSG5174)', '좌표정보Y(EPSG5174)'], keep='first')
filtered_df
```

```python
filtered_df['사업장명']
```

```python
## '소재지전체주소' 컬럼 정리 (번지수 까지만 남기기)
filtered_df['소재지전체주소'] = filtered_df['소재지전체주소'].apply(lambda x: ' '.join(x.split()[:4]))

## '상세보기주소' 컬럼 추가 (NaN 값으로 채움)
filtered_df['상세보기url'] = np.nan

# 결과 출력
filtered_df
```

```python
## 상세보기주소 크롤링 (ver. 250529)
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import csv
import pandas as pd
import numpy as np
import re
from googletrans import Translator

import warnings
warnings.filterwarnings('ignore')   # 경고문 무시

url = 'https://map.kakao.com/'
driver = webdriver.Chrome()
driver.get(url)

result_url_list = []

## 모든 한글 음절이 한 번 이상 연속으로 나타나는 패턴 정의
korean_pattern = re.compile(r'[\uAC00-\uD7A3]+')

## 상호명 전처리 함수
def storename_preprocess(text):
    if not text.strip():    # 문자열이 없는 공백인 경우
        # print("입력된 문자열이 없습니다!")
        cleaned_text = ['상호명없음']
        return cleaned_text    
    
    ## 한글(패턴) 존재여부 확인후 없으면 단순번역(한글)+원문 
    korean_pattern = re.compile(r'[\uAC00-\uD7A3]+')
    has_korean = korean_pattern.search(text)

    if not has_korean:
        translator = Translator()
        text = (translator.translate(text, dest='ko').text) + ' ' + text

    ## 본격 상호명 전처리
    processed_text = re.sub(r'[(){}\[\]<>_&\(주\)]', ' ', text)      # 모든 종류의 괄호와 밑줄, '&', '(주)'는 빈칸으로 대체
    cleaned_text = re.sub(r'[^\w\s\uAC00-\uD7A3]', '', processed_text)  # 나머지 특수문자 모두 삭제 (알파벳, 한글, 숫자, 공백만 남김)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()        # 연속된 빈칸을 최소 1칸만 남김

    token_pattern = re.compile(r'[\uAC00-\uD7A3]+|[a-zA-Z]+|\d+')   # 연속된 한글, 연속된 영문, 연속된 숫자를 각각의 '단어'로 간주하여 추출
    tokens = token_pattern.findall(cleaned_text)    # 패턴과 일치하는 모든 부분을 리스트로 저장
    
    return tokens   # 저장된 리스트를 반환


## 검색 결과('장소')에서 상호에 해당하는 '상세보기' 링크 가져오는 함수2(좀 더 정확도 높음)
def find_moreview_url(serch_shop_name_list, soup):    
    
    ## 입력 리스트 요소 중 하나라도 대상 문자열에 포함되어 있는지 확인
    def contains_any_word(word_list, target_string):        
        for word in word_list:
            if word in target_string:
                return True  # 일치하는 부분을 찾으면 즉시 True 반환
        return False # 리스트의 모든 단어를 확인했지만 일치하는 부분이 없으면 False 반환

    ## soup 객체 내 검색된 '장소' 리스트 영역
    place_list = soup.find("ul", class_="placelist", id="info.search.place.list")

    ## '장소' 리스트 중 목표 상호명과 일치하는 것의 '더보기' 링크 찾아 반환
    for link_name_link in place_list.find_all('a', class_="link_name"):

        target_link_name = link_name_link.string    # 검색 결과 내 상호명

        if not contains_any_word(serch_shop_name_list, target_link_name):        
            continue        
        else:
            next_element = link_name_link.find_next('a', class_="moreview")
            moreview_url = next_element['href']
            return moreview_url

    return np.nan # 조건을 만족하는 '더보기' 링크를 찾지 못한 경우

## 크롤링 반복작업

for i in range(len(filtered_df)):
    serch_shop_name_list = storename_preprocess(filtered_df['사업장명'].iloc[i])
    serch_shop_addres = filtered_df['소재지전체주소'].iloc[i]       

    search_area = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')   # 검색창 선택
    search_area.clear()     # 검색창 비우기
    search_area.send_keys(serch_shop_addres)    ## 검색어 넣고 엔터
    time.sleep(0.2)     ## 로딩 대기 시간
    driver.find_element(By.XPATH, '//*[@id="search.keyword.submit"]').send_keys(Keys.ENTER)
    time.sleep(2)       ## 로딩 대기 시간

    ## 검색 결과에서 '장소' 버튼 클릭
    driver.find_element(By.XPATH, '//*[@id="info.main.options"]/li[2]/a').send_keys(Keys.ENTER)
    time.sleep(0.2)

    ## 페이지 내용 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
        
    moreview_url = find_moreview_url(serch_shop_name_list, soup)
    
    print('no.', i+1, ' 상호명=', filtered_df['사업장명'].iloc[i], ' /  상세보기URL=', moreview_url)
    result_url_list.append(moreview_url)
    
    # filtered_df['상세보기url'].iloc[i] = moreview_url    

print('**크롤링 완료**')
driver.quit()
```
